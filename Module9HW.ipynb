## Flights Questions
### Use flights_data

```python
#Question 1 How many flights were there from JFK to SLC? Int

#flight origin is JFK and destination is SLC
#use sum() to count the number of rows
q_1 = sum((flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC'))
```

```python
#Question 2 How many airlines fly to SLC? Should be int

#filter: destination is SLC
#'carrier' is the column we want to extract, so we make an array of all values fitting the condition
#nunique() to count number of unqiue rows in 'carrier' column
q_2 = flights_data.loc[flights_data['dest'] == 'SLC', 'carrier'].nunique()
```

```python
#Question 3 What is the average arrival delay for flights to RDU? float

#filter: destination is RDU
#making an array of all arr_delay rows
#using .mean() to find average
q_3 = flights_data.loc[flights_data['dest'] == 'RDU', 'arr_delay'].mean()
```

```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float


#numerator filter: orgin is JFK or LGA, destination is SEA
#denominator filter: destination is Seattle (regardless of origin)
#sum(num)/sum(denom) = %of flights to SEA from LGA/JFK
q_4 = sum(((flights_data['origin'] == 'JFK') | (flights_data['origin'] == 'LGA')) & (flights_data['dest'] == 'SEA'))/sum(flights_data['dest'] == 'SEA') 
```

```python
#Prep for question 5..

#please make date a column. Preferred format is 2013/1/1 (y/m/d)

#using pd.to_datetime to create a date based of year,month, and day columns
flights_data['date'] = pd.to_datetime(flights_data[['year','month','day']])
```

```python
#Question 5 Which date has the largest average depature delay? Pd slice with date and float
q_5_1 = flights_data.groupby('date').agg({'dep_delay' : ['mean']})
q_5_2 = q_5_1.sort_values(by = [('dep_delay', 'mean')], ascending = False)
q_5 = q_5_2[:1]
```

```python
#Question 6 Which date has the largest average arrival delay? pd slice with date and float

#group by date, aggregate by max arrival delay
q_6_1 = flights_data.groupby('date').agg({'arr_delay' : ['max']})

#sort values descending (largest delay on top)
q_6_2 = q_6_1.sort_values(by = [('arr_delay', 'max')], ascending = False)

#only show top row using slice (largest delay)
q_6 = q_6_2[:1]
```

```python
#Q7 prep: create 'speed' column

#speed = distance/airtime
flights_data['speed'] = flights_data['distance']/flights_data['air_time']
```

```python
#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed

#filters: orgin is LGA or JFL
#sort by speed descending (fastest on top)
q_7_1 = flights_data[(flights_data['origin']=='LGA')|(flights_data['origin']=='JFK')].sort_values(by = ['speed'], ascending = False)

#filter down to tailnum, speed. Only show top row using slice
q_7 = q_7_1[['tailnum','speed']][:1]
```
